# Graph Neural Network (GNN)

## Purpose
Graphs are dynamic, non-structed data that standard DNN cannot handle. Furthermore, when graphs become huge, they become hard to interprete by a humain being.

   $\rightarrow$ Solution: Graph Neural Network

## Defintion & tasks

Graph Neural Networks (GNNs) are a variant of Deep Neural Network (DNN) that can be used to train on graphs. GNN are applied to node-based tasks (Node classification, Link prediction, Community detection, ...), but also graph-based tasks (Graph classification, Graph Generation ...).

## Layers & Architectures of GNN

### - Layers

#### 1- Graph Filtering Layer
Inputs: 
- Adjacency matrix A of size N * N, where N is the number of nodes
- Input Features F(if) of size N * d(if), where d(if) is the dimension of the input features.
        
Outputs: 
- Output Features F(of) of size N * d(of), where d(of) is the dimension of the output features.

Role: Refine the node features while maintaining the graph structure.

Use cases: Node-based tasks, Graph-based tasks.

#### 2- Graph Pooling Layer
Inputs: 
- Adjacency matrix A(ip) of size N(ip) * N(ip), where N(ip) is the number of nodes of the input graph.
- Input Features F(ip) of size N(ip) * d(ip), where d(ip) is the dimension of the input features.
        
Outputs: 
- Adjacency matrix A(op) of size N(op) * N(op), where N(op) is the number of nodes of the output graph.
- Output Features F(of) of size N(op) * d(op), where d(op) is the dimension of the output features.

Role: Aggregate the node features to obtain a new graph structure (coarsened graph).

Use cases: Graph-based tasks.

### - Architectures

#### 1- Architecture for Node-based tasks
Used Layers: 
- Graph Filtering Layer $h(A, F)$, where A is the adjacency matrix and F represents the features.
- Activation Layer $a(F)$, where F represents the features.
        
Global architecture functionning: 

At first, the input features are passed to a graph filter which refines them. The refined features are passed into an activation layer (ReLu, ..). The resulting features are then passed into another graph filter, and so on. The last output features may be passed to specific additionnal layers depending on the task. As follows, the recursive definition of this process:

$$
\begin{cases}
F_i = h(A, a_{i-1}(F_{i-1})), \text {for i = 1..k where k is the number of graph filters applied}\\
a_0 = \text {Identity function}\\
F_0 = F: \text {Input Features}
\end{cases}
$$

#### 2- Architecture for Graph-based tasks
Used Layers: 
- Graph Filtering Layer $h(A, F)$, where A is the adjacency matrix and F represents the features.
- Activation Layer $a(F)$, where F represents the features.
- Graph Pooling Layer $p(A, F)$, where A is the adjacency matrix and F represents the features.

GNN Block architecture: 

A block consists of a sequence of graph filters and activation layers used to refine the input graph features. The resulting features are then passed into a pooling layer which aggregate the input graph into a coarsened graph. As follows, the recursive definition of a GNN block for graph-based tasks: 
$$
\begin{cases}
F_i = h(A_{in}, a_{i-1}(F_{i-1})), \text {for i = 1..k where k is the number of graph filters applied}\\
A_{out}, F_{out} = p(A_k, F_k))\\
a_0 = \text {Identity function}\\
F_0 = F_{in}: \text {Input Features}
\end{cases}
$$

<img src=attachment:e8392391-9551-49cd-8e18-6eb59c86b7f9.png style="width: 50%; display: block; margin-left: auto; margin-right: auto;"/>

Global architecture:

It consists of connecting several GNN blocks together to form a sequence where the input of the current block is the output of the previous one. Each block takes a graph as input and then outputs a coarsened graph. As follows, the recursive definition of this process:

$$
\begin{cases}
A_i, F_i = Block_i(A_{i-1}, F_{i-1})), \text {for i = 1..L where L is the number of blocks}\\
A_0 = A: \text {Adjacency matrix of the original graph}\\
F_0 = F: \text {Input Features}
\end{cases}
$$

## GCN with Pytorch-Geometric

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
import torch
from torch_geometric.datasets import Planetoid
from torch_geometric.transforms import NormalizeFeatures

dataset = Planetoid(root='data/Planetoid', name='Cora', transform=NormalizeFeatures())

print(f'Dataset: {dataset}:')
print('======================')
print(f'Number of graphs: {len(dataset)}')
print(f'Number of features: {dataset.num_features}')
print(f'Number of classes: {dataset.num_classes}')

data = dataset[0]  # Get the first graph object.
print(data)
```

```python
from torch_geometric.nn import Sequential, GCNConv
import torch.nn.functional as F

class GCN(torch.nn.Module):
    def __init__(self, hidden_channels):
        super().__init__()
        torch.manual_seed(1234567)
        self.conv1 = GCNConv(dataset.num_features, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, dataset.num_classes)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = x.relu()
        x = F.dropout(x, p=0.5, training=self.training)
        x = self.conv2(x, edge_index)
        return x

model = GCN(16)
print(model)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

def visualize(h, color):
    z = TSNE(n_components=2).fit_transform(h.detach().cpu().numpy())

    plt.figure(figsize=(10,10))
    plt.scatter(z[:, 0], z[:, 1], s=70, c=color, cmap="Set2")
    plt.show()
```

```python
model.eval()

out = model(data.x, data.edge_index)
visualize(out, color=data.y)
```

```python
model = GCN(hidden_channels=16)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
criterion = torch.nn.CrossEntropyLoss()

def train():
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = criterion(out[data.train_mask], data.y[data.train_mask])
    loss.backward()
    optimizer.step()
    return loss

def test():
    model.eval()
    out = model(data.x, data.edge_index)
    pred = out.argmax(dim=1)
    test_correct = pred[data.test_mask] == data.y[data.test_mask]
    test_acc = int(test_correct.sum()) / int(data.test_mask.sum())
    return test_acc


for epoch in range(100):
    loss = train()
print('Training complete.')
```

```python
test_acc = test()
print(f'Test Accuracy: {test_acc:.4f}')
```

```python
model.eval()
out = model(data.x, data.edge_index)
visualize(out, color=data.y)
```
